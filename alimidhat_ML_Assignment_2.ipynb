```python
# Generalized function that can handle multi-calss porblems
def gini_idx(lbls):
  # count number of labels
  n_total = len(lbls)
  if n_total == 0:
    return 0 # No samples in this group (avoid division by zero)

  # calcuate the count of each label
  unique_lbls , counts = np.unique(lbls, return_counts=True)

  gini = 1 - sum( (count/n_total)**2 for count in counts)
  return gini
```

##1.3

```python
def cart_loss(lbls, feature, threshold):
  # Segment data points
  left_data = [] # less than threshold
  right_data = [] # more than threshold

  for i in range(len(feature)):
    if feature[i] <= threshold:
      left_data.append(lbls[i])
    else:
      right_data.append(lbls[i])
  # calculate gini for both data points
  gini_left = gini_idx(left_data)
  gini_right = gini_idx(right_data)

  loss = gini_idx(left_data)*(len(left_data)/len(feature)) + gini_idx(right_data) * (len(right_data)/len(feature))
  return gini_left, gini_right, loss
```

##1.4

```python
# Takes only one feature per time
def decision_stumper(lbls, feature, step_size=1):
    # takes the range with respect to the step size and the highest feature value
    range = np.arange(min(feature), max(feature)+step_size, step_size)
    threshold = 0
    losses = {}
    for iteration, i in enumerate(range):
      g_left, g_right, loss = cart_loss(lbls, feature, threshold)
      # dict to holds each threshold with the respective loss
      losses[str(threshold)] = loss
      if len(range) < 100 and iteration%10==0:
          print(f"iteration: {iteration}\t threshold: {threshold:.2f}\t Loss: {loss:.4f}\t gini_value(left, right): {(g_left, g_right)}")
      elif len(range) >= 100 and iteration%50==0:
          print(f"iteration: {iteration}\t threshold: {threshold:.2f}\t Loss: {loss:.4f}\t gini_value(left, right): {(g_left, g_right)}")
      threshold += step_size
    return losses
```

###Decision stumper in Action

```python
feature_1 = "detajj"
feature_2 = "massjj"
labels = "sample"

train_f1 = train_dataset[feature_1]
train_f2 = train_df[feature_2]
train_lbl = train_df[labels]
losses_f1 = decision_stumper(train_lbl, train_f1, step_size=0.1)
```

> *This "overtraining check" plot compares the classifier's performance on training and test samples for both signal and background. It helps ensure the model isn't overfitting 'overtraining' and can generalize well to unseen data.*

##2.2

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=2)
clf.fit(train_df[["detajj", "massjj"]], train_df["sample"])
```

```python
def plot_overtraining_check(classifier, X_train, y_train, X_test, y_test, positive_class=1, title="Overtraining Check"):
    # Compute classifier scores for train and test sets
    train_scores = classifier.predict_proba(X_train)[:, positive_class]
    test_scores = classifier.predict_proba(X_test)[:, positive_class]

    # Separate signal and background scores
    train_signal = train_scores[y_train == positive_class]
    train_background = train_scores[y_train != positive_class]
    test_signal = test_scores[y_test == positive_class]
    test_background = test_scores[y_test != positive_class]

    # Bins of the histogram
    bins = np.linspace(0, 1, 50)

    # Plot test sample histograms (signal, background)
    plt.hist(
        test_signal, bins=bins, density=True, alpha=0.5, label="Signal (test sample)", color="skyblue"
    )
    plt.hist(
        test_background, bins=bins, density=True, alpha=0.5, label="Background (test sample)", color="lightcoral"
    )

    # Compute normalized weights and errors for training sets
    train_signal_hist, _ = np.histogram(train_signal, bins=bins, density=True)
    train_background_hist, _ = np.histogram(train_background, bins=bins, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    train_signal_err = np.sqrt(train_signal_hist / len(train_signal))
    train_background_err = np.sqrt(train_background_hist / len(train_background))

    # Add training set data as error bars
    plt.errorbar(
        bin_centers, train_signal_hist, yerr=train_signal_err, fmt='.', label="Signal (train sample)", color="blue"
    )
    plt.errorbar(
        bin_centers, train_background_hist, yerr=train_background_err, fmt='.', label="Background (train sample)", color="red"
    )

    plt.title(title)
    plt.xlabel("Classifier Scores")
    plt.ylabel("Normalized Units")
    plt.legend(loc="upper right") # fix it at top right location
    plt.grid()
    plt.show()
```

```python
plot_overtraining_check(clf, train_df[["detajj", "massjj"]], train_df["sample"], test_df[["detajj", "massjj"]], test_df["sample"], title="Overtraining Check, Decision Tree with max depth 2")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(train_df[["detajj", "massjj"]], train_df["sample"])

plot_overtraining_check(rf_clf, train_df[["detajj", "massjj"]], train_df["sample"], test_df[["detajj", "massjj"]], test_df["sample"], title="Overtraining Check, RandomForestClassifier with 100 estimators")
```

>the model is clearly overfitting. It performs very well on the training data, where signal and background are almost perfectly separated, but struggles with the test data. There’s significant overlap between the test signal and background, indicating the model has memorized the training data instead of learning patterns that generalize to new data.

```python
rf_clf_2 = RandomForestClassifier(max_leaf_nodes=32, random_state=42)
rf_clf_2.fit(train_df[["detajj", "massjj"]], train_df["sample"])

plot_overtraining_check(rf_clf_2, train_df[["detajj", "massjj"]], train_df["sample"], test_df[["detajj", "massjj"]], test_df["sample"], title="Overtraining Check: RandomForestClassifier max_leaf_nodes = 32")
```

>*In the second plot, the model shows less overfitting compared to the first case. The test data aligns better with the training data, especially for the signal. While there’s still some overlap between the signal and background in the test set, limiting the complexity of the model with `max_leaf_nodes=32` has helped it generalize better, reducing the overfitting seen in the first case*
